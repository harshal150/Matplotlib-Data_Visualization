```python
x = ['a','b','c','d','e']
y = np.random.rand(5)
```

```python
y
```

```python
plt.bar(x,y)
plt.xlabel("this is x data")
plt.ylabel("this is y data")
plt.title("graph title !!")
```

```python
plt.barh(x,y)
plt.xlabel("this is x data")
plt.ylabel("this is y data")
plt.title("graph title !!")        # barh fun tranfer x & y !! x become y and y become x
```

```python
plt.plot(x,y,"--r") 
plt.xlabel("this is x data")
plt.ylabel("this is y data")
plt.title("graph title !!")
plt.show()
```
